### 3. 데이터 분석

#### 3-1 총지출액 계산

```python
spend = 0
for row in data:
    if row[-1] == '전표매입' :
        spend += int(row[-3])

print(spend)
```

#### 3-2 월별 지출액 그래프 그리기

```python
# 빈 리스트 생성
quarter4 = [0, 0, 0]
    
for row in data:
    if row[-1] == "전표매입" :
        # row를 '-' 단위로 나누고, 그중 두번째로 나오는 것 
        month, pay = int(row[0].split('-')[1]), int(row[-3])
        idx = month - 10
        quarter4[idx] += pay

print(quarter4)

plt.rc('font', family = 'Gulim')
plt.title("4분기 지출 내역", pad = 20)
plt.bar(['10월', '11월', '12월'], quarter4, color = 'yellowgreen')
plt.show()
```

#### 3-3 지출 top 20

```python
spend = {}

for row in data:
    if row[-1] == "전표매입" :
        store, pay = row[-4], int(row[-3])
        
        # 없으면 새로 추가, 있으면 누적
        if store not in spend.keys():
            spend[store] = pay
        else :
            spend[store] += pay

#spend

top20 = sorted(spend.items(), key = operator.itemgetter(1), # 값 기준 정렬
              reverse = True)[:20]
top20_place, top20_sum = [], []

for i in top20:
    top20_place.append(i[0])
    top20_sum.append(i[1])
    
# 그래프 작성
plt.title("4분기 매출 Top-20")
plt.barh(top20_place, top20_sum, color = 'crimson')
plt.show()
```

### Self Check

#### 1. 취소거래의 총액

```python
cancel = 0

for row in data:
    if row[-1] != "전표매입" :
        cancel += int(row[-3])
print(cancel)
```

#### 2. 백화점 지출 금액 꺾은선 그래프

```python
depart = [0, 0, 0]
    
for row in data:
    if "백화점" in row[-4]:
        depart[int(row[0].split('-')[1]) - 10] += int(row[-3])

print(depart)

plt.rc('font', family = 'Gulim')
plt.title("백화점 지출 내역", pad = 15)
plt.plot(['10월', '11월', '12월'], depart, color = 'brown')
plt.show()
```
