### 2. 꺾은선 그래프
시계열 데이터를 표현하기 유용 

```python
sales = [1, 5, 9, 18, 13]
month = ["Jan", "Apr", "Jun", "Sep", "Dec"]

# 한글로 그래프 제목이나 축제목을 입력하려면, 별도의 옵션을 지정해야 된다.
plt.rc('font', family = 'Malgun Gothic') # Mac은 AppleGothic

# 명령어의 괄호 안에서 shift + tab을 누르면 param 목록을 확인할 수 있다.

plt.figure(dpi = 300, # 해상도
           figsize = (10, 8)) # 그래프 크기
plt.title("Monthly Sales Trend")
plt.plot(month, sales, 
         color = 'g', # 그래프 색 지정
         label = "단위 : 개수" # 범례 설정
        )

# 범례 표시
plt.legend(loc= 'best' # 범례 위치
          ) 

# 축 제목
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.show()
```

```python
def ran_num(x, y, z):
    
    test = []

    for i in range(x):
        a = np.random.randint(y, z) 
        test.append(a)  
    
    return test
```

#### 2-1 겹쳐그리기

```python
nations = ['korea', 'usa', 'japan', 'china', 'india']

football = ran_num(5, 10, 100)
baseball = ran_num(5, 10, 100)

plt.plot(nations, football, label = "축구", color = 'red')
plt.plot(nations, baseball, label = "야구", color = 'skyblue')
plt.legend(loc = 'best')
plt.show()
```

### 2. 막대그래프

```python
em = []

for i in range(10):
    hire = np.random.rand(10) * 100
    em.append(round(hire[i],0))
```

```python
month = np.arange(1,11); month
```

```python
plt.title('monthly employment')
plt.barh(month, em, color = 'Orange') # bar, barh 상황에 맞게 활용
plt.show()
```

#### 2-2 겹쳐그리기

```python
plt.rcParams['axes.unicode_minus'] = False # 음수부호 인식

for i in range(len(football)):
    football[i] = -football[i] 

plt.title('국가별 스포츠 점수')
plt.barh(nations, football, label = "축구", color = "yellowgreen")
plt.barh(nations, baseball, label = "야구", color = "purple")
plt.legend()
plt.show()
```

### 4. 원그래프

### 5. 산점도

```python
math = ran_num(15, 30, 100)
sci = ran_num(15, 30, 100)
plt.scatter(sci, math)
plt.xlabel('과학 점수')
plt.ylabel('수학 점수')
plt.show()
```

#### 5-1 컬러맵 적용

```python
pizza = ran_num(200, 100, 200)
chicken = ran_num(200, 100, 200)

plt.title('Relation between pizza & chicken')
plt.scatter(pizza, chicken, 
            s = ran_num(200, 30, 500), # 점 크기, x, y와 길이가 같아야 함
           c = pizza, # pizza값에 따라 색이 다르게 나옴.
           cmap = 'gnuplot2')
plt.xlabel('피자')
plt.ylabel('치킨')
plt.colorbar(label = "피자")
plt.show()
```

#### 5-2 겹쳐그리기

```python
pizza1 = ran_num(200, 100, 200)
chicken1 = ran_num(200, 100, 200)
pizza2 = ran_num(200, 100, 200)
chicken2 = ran_num(200, 100, 200)

plt.title('서울, 부산의 피자, 치킨 점포수')
plt.scatter(pizza1, chicken1, 
            s = ran_num(200, 30, 150), # 점 크기, x, y와 길이가 같아야 함
           color = 'indigo',
           alpha = 0.5, # 투명도
           label = "Seoul")
plt.scatter(pizza2, chicken2, 
            s = ran_num(200, 30, 150), # 점 크기, x, y와 길이가 같아야 함
           color = 'Gray',
           alpha = 0.5, # 투명도
           label = "Busan")
plt.xlabel('피자')
plt.ylabel('치킨')
plt.legend()
plt.colorbar(label = "피자")
plt.show()
```

### 7. Self-Check

```python
weekday = ran_num(8,1, 25)
weekend = ran_num(8,1, 20)
```

#### 7-1 겹쳐진 꺾은선 그래프 

```python
plt.title("주중/주말 자동차 통행량 비교")
plt.plot(range(8), weekday, color = 'crimson', label = '주중 통행량')
plt.plot(range(8), weekend, color = 'royalblue', label = '주말 통행량')
plt.legend(loc = 'upper left')
plt.show()
```

#### 7-2 겹쳐진 막대 그래프

```python
for i in range(len(weekday)):
    weekday[i] = -weekday[i]
    
plt.title("주중/주말 자동차 통행량 비교")
plt.barh(range(8), weekday, color = 'crimson', label = '주중 통행량')
plt.barh(range(8), weekend, color = 'royalblue', label = '주말 통행량')
plt.legend(loc = 'lower left')
plt.show()
```
